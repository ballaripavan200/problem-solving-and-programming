## problem 1:
 
 plotting 2D graphs using matplotib

```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

plt.plot([1,2,3,4],[1,8,27,64])
plt.xlabel('time')
plt.ylabel('distance')
plt.show()
```

```python
import numpy as np

t=np.arange(0.0,2.0,0.1)
v=1+np.sin(2*np.pi*t)
plt.xlabel('time')
plt.ylabel('voltage')
plt.title('sine wave')
plt.plot(t,v)
plt.grid(True)
plt.savefig('figures/time-voltage.png')
plt.show()
```

## problem 4 :

generate a scatter plot of 50 random numbers

```python
n=50
x = np.random.rand(n)
y = np.random.rand(n)
colourlist = np.random.rand(n)
arealist = np.pi * (15*np.random.rand(50)**2)
plt.scatter(x,y,s=arealist,c=colourlist,alpha=1)
plt.savefig('figures/scatter')
plt.show()
```
