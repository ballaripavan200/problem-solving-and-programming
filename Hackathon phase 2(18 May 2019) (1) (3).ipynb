## Problem 1:
Using the Matplotlib library, generate a Pie Chart to categorise student marks data in a CSV file into Distinction, First Class, Second Class, Third Class and Failed Categories(based on percentages). Explode the Distinction category.

* Generate a CSV containing marks of 2000 students

* From the CSV, generate the Pie Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
def distinction(n):
    return n>=80
def firstclass(n):
    return 60<=n<80
def secondclass(n):
    return 50<=n<60

df = pd.read_csv('Datafiles/marks.csv',usecols=['marks'])
marks = list(map(int,df.values))
dis = (sum(map(distinction,marks)))/2000
fi = (sum(map(firstclass,marks)))/2000
s = (sum(map(secondclass,marks)))/2000
t = (sum(1 for i in marks if i>=40 and i<50))/2000
fail = (sum(1 for i in marks if i<40))/2000
labels = 'Distinction','Firstclass','Secondclass','Thirdclass','Fail'
sizes = [dis*100,fi*100,s*100,t*100,fail*100]
colors = ['blue','green','yellow','red','pink']
explode = (0.1,0,0,0,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1f%%',shadow=True,startangle=90)
plt.show()
```

#### a) Pie Chart to categorise all app categories based on count.

```python
import pandas as pd
df=pd.read_csv('DataFiles/AppleStore.csv')
df2=pd.read_csv('DataFiles/AppleStore.csv',usecols=['prime_genre'])
a=df2.values
u=[]
for x in df2.values:
    if x not in u:
        u.append(x)
c=[]
for x in u:
    count=0
    for i in range(0,len(a)):
        if a[i]==x:
            count=count+1
    c.append(count)
sizes=c
plt.pie(sizes,labels=u,startangle=90)
plt.show()
```

#### b)For the Sports Category, generate a bar graph to show average rating for free vs paid apps.

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('DataFiles/AppleStore.csv')
df2=pd.read_csv('DataFiles/AppleStore.csv',usecols=['prime_genre'])
a=df2.values

count1=count2=free=paid=0

for i in range(0,len(a)):
    if a[i][0]=='Sports':
        if df.loc[i][5]==0.00:
            free=free+df.loc[i][8]
            count1=count1+1
        else:
            paid=paid+df.loc[i][8]
            count2=count2+1
free=free/count1
paid=paid/count2
x=['free','paid']
y=[free,paid]
width=1/2
plt.xlabel('free or paid')
plt.ylabel('average rating')
plt.title('free vs paid for sports apps')
plt.bar(x,y,width,color='green')
plt.show()
```
